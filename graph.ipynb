```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv('timemap.txt', sep=' ', names=['p', 'h', 'correct_time', 'custom_time'])

# Get unique 'p' values
p_values = data['p'].unique()

# Manually specify colors for better differentiation
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# Create a larger plot
plt.figure(figsize=(15, 10))

for i, p in enumerate(p_values):
    subset = data[data['p'] == p]

    # Plot 'correct_time' and 'custom_time' for the current 'p' value with different color
    plt.plot(subset['h'], subset['correct_time'], marker='o', color=colors[i%len(colors)], label=f'Correct Time (p={p})')
    plt.plot(subset['h'], subset['custom_time'], marker='o', linestyle='--', color=colors[i%len(colors)], label=f'Custom Time (p={p})')

plt.xscale('log')
plt.yscale('log')

plt.title('Plot of h vs Time')
plt.xlabel('h')
plt.ylabel('Time')

# Configure major and minor gridlines
plt.grid( which='major', color='gray', linestyle='-')
plt.grid( which='minor', color='gray', linestyle='--', alpha=0.5)

plt.gca().set_xticks([10**i for i in range(int(np.log10(data['h'].min())), 1+int(np.log10(data['h'].max())))], minor=True)
plt.gca().set_yticks([10**i for i in range(int(np.log10(data[['correct_time', 'custom_time']].min().min())), 1+int(np.log10(data[['correct_time', 'custom_time']].max().max())))], minor=True)

# Move the legend outside of the plot
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('timemap.txt', sep=' ', names=['p', 'h', 'correct_time', 'custom_time'])

# Get unique 'p' values
p_values = data['p'].unique()

# Create subplots
fig, axs = plt.subplots(len(p_values), 1, figsize=(10, 6*len(p_values)))

for i, p in enumerate(p_values):
    subset = data[data['p'] == p]

    # Plot 'correct_time' and 'custom_time' for the current 'p' value
    axs[i].plot(subset['h'], subset['correct_time'], marker='o', color='blue', label='Correct Time')
    axs[i].plot(subset['h'], subset['custom_time'], marker='o', color='red', label='Custom Time')

    axs[i].set_xscale('log')
    axs[i].set_yscale('log')

    axs[i].set_title(f'Plot of h vs Time for p={p}')
    axs[i].set_xlabel('h')
    axs[i].set_ylabel('Time')
    axs[i].legend(loc='best')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
